```python
import soundfile as sf
from pathlib import Path
import librosa
from os import (listdir,remove)
import pandas as pd
import numpy as np
```

```python
base_dir=Path("E:\Cornell_Bird_Call_Noise")
```

```python
def convert_wav():
    folders=[f for f in listdir(base_dir)]
    for folder in folders:
        sub_dir=base_dir / folder
        files=[f for f in listdir(sub_dir)]
        for f in files:
            ext=f.split('.')[1]
            if ext!='wav':
                print("converting file {} to wav".format(f))
                y, _ = librosa.load(
                    sub_dir / f ,
                    sr=32000, 
                    mono=True, 
                    res_type="kaiser_fast"
                )
                remove(sub_dir / f)
                f_list=f.split('.')
                f_list[1]=".wav"
                f=''.join(f_list)
                sf.write(sub_dir / f, y, samplerate=32000)
            else :
                y, sr = sf.read(sub_dir / f)
                if sr!=32000:
                    print("resampling file {} to 32k".format(f))
                    remove(sub_dir / f)
                    out = librosa.resample(np.asfortranarray(y), sr, 32000, res_type='kaiser_fast')
                    f_list=f.split('.')
                    f_list[0]=f_list[0]+'_32k'+'.'
                    f=''.join(f_list)
                    sf.write(sub_dir / f, out, samplerate=32000)

convert_wav()
         
```

```python
def convert_5_seconds():
    folders=[f for f in listdir(base_dir)]
    dit={"file_name":[],"type":[]}
    final_dir=base_dir / "10_sec_clip"
    if  not final_dir.exists():
        final_dir.mkdir(parents=True, exist_ok=True)
    for folder in folders:
        sub_dir=base_dir / folder
        files=[f for f in listdir(sub_dir)]
        count=-1
        for f in files:
            y, sr = sf.read(sub_dir / f)
            print(f)
            print(sr)
            print(len(y))
            count+=1
            for i in range(0,len(y),10*sr):
                f_list=f.split('.')
                f_list[0]=str(folder)+"_"+str(count)+"_"+str(i//(10*sr))+'.'
                f_exit=''.join(f_list)
                dit["file_name"].append(f_exit)
                dit["type"].append(str(folder))
                ##sf.write(final_dir / f_exit,y[i:i+10*sr],samplerate=32000)
    dit=pd.DataFrame(dit)
    dit.to_csv(final_dir / "df.csv")
    print("csv generated")
    return dit
                
df=convert_5_seconds()           
```
